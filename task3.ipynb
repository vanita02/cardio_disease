```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('explored_dataset.csv')
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(['cardio','age','Unnamed: 0','id'],axis=1)
y = data['cardio']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print('train size : ',x_train.shape)
print('test size : ',x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

num_cols = [
    'calculated_age', 'height', 'weight',
    'ap_hi', 'ap_lo', 'bmi'
]

x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

```python
#sigmoid function

def sigmoid(z):
    return 1/(1+np.exp(-z))


#logistic_regression function

def logistic_regression(X,y, learning_rate=0.1,epochs=20000):
    m,n = X.shape
    weights = np.zeros(n)
    bias=0

    for i in range(epochs):
        z = np.dot(X,weights) + bias

        prediction = sigmoid(z)

        lr = 0.1 / np.sqrt(i + 1)

        dw = (1/m) * np.dot(X.T,(prediction-y))
        db = (1/m) * np.sum(prediction-y)

        weights -= learning_rate * dw
        bias -= learning_rate * db
    return weights, bias

def predict(X, weights, bias):
    y_pred = sigmoid(np.dot(X, weights) + bias)
    return (y_pred>=0.5).astype(int)

#accuracy function

def accuracy(y_true,y_pred):
    return np.sum(y_true == y_pred) / len(y_true)
```

```python
# train the model

weights, bias = logistic_regression(x_train, y_train)
```

```python
# prediction 

y_pred = predict(x_test,weights,bias)
```

```python
acc = accuracy(y_test, y_pred)
print('Accuracy : ',acc)
```
